```python



#Question 1: Car Sales Analysis Dashboard Problem Statement A car showroom wants to analyze the sales performance of different car brands.

import pandas as pd
import matplotlib.pyplot as plt
# Dataset
data = {
    "Brand": ["Toyota", "Honda", "Hyundai", "Tata", "Mahindra", "Kia"],
    "Sales": [120, 95, 110, 85, 100, 90],
    "Price": [12, 10, 11, 9, 14, 13],
    "Mileage": [18, 20, 19, 22, 17, 18]
}

df = pd.DataFrame(data)

# Create Dashboard
plt.figure(figsize=(12,8))

# Graph 1: Bar Plot
plt.subplot(2,2,1)
plt.bar(df["Brand"], df["Sales"], color=plt.cm.Set2.colors)
plt.title("Car Sales")
plt.xlabel("Brand")
plt.ylabel("Sales")

# Graph 2: Scatter Plot
plt.subplot(2,2,2)
plt.scatter(df["Price"], df["Sales"], s=150, color="green")
plt.title("Price vs Sales")
plt.xlabel("Price")
plt.ylabel("Sales")

# Graph 3: Histogram
plt.subplot(2,2,3)
plt.hist(df["Mileage"], bins=6, color="orange")
plt.title("Mileage Distribution")
plt.xlabel("Mileage")
plt.ylabel("Frequency")

# Graph 4: Box Plot
plt.subplot(2,2,4)
plt.boxplot(df["Sales"], patch_artist=True,
            boxprops=dict(facecolor="skyblue"))
plt.title("Sales Outlier Detection")
plt.ylabel("Sales")

plt.tight_layout()
plt.show()
```

```python
#Question 2: Hospital Patient Analysis Dashboard Problem Statement A hospital wants to analyze patient information using different graphs.

import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    "Patient": ["P1","P2","P3","P4","P5","P6","P7","P8"],
    "Age": [25,40,35,60,45,30,55,50],
    "BloodPressure": [120,140,130,160,150,125,155,145],
    "Sugar": [90,110,105,180,150,95,170,140]
}

df = pd.DataFrame(data)

# Create Dashboard
plt.figure(figsize=(12,8))

# Graph 1: Line Plot
plt.subplot(2,2,1)
plt.plot(df["Patient"], df["BloodPressure"],
         marker="o", color="blue")
plt.title("Blood Pressure")
plt.xlabel("Patient")
plt.ylabel("Blood Pressure")

# Graph 2: Scatter Plot
plt.subplot(2,2,2)
plt.scatter(df["Age"], df["Sugar"],
            color="red", s=120)
plt.title("Age vs Sugar")
plt.xlabel("Age")
plt.ylabel("Sugar")

# Graph 3: Histogram
plt.subplot(2,2,3)
plt.hist(df["BloodPressure"],
         bins=5,
         color="purple")
plt.title("Blood Pressure Distribution")
plt.xlabel("Blood Pressure")
plt.ylabel("Frequency")

# Graph 4: Box Plot
plt.subplot(2,2,4)
plt.boxplot(df["Sugar"],
            patch_artist=True,
            boxprops=dict(facecolor="lightgreen"))
plt.title("Sugar Outliers")
plt.ylabel("Sugar")

plt.tight_layout()
plt.show()
```
