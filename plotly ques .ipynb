```python
import pandas as pd

df = pd.DataFrame({
    "Order_ID": range(101, 121),
    "Product": [
        "Laptop","Mobile","Tablet","Laptop","Mobile",
        "Tablet","Laptop","Mobile","Tablet","Laptop",
        "Mobile","Tablet","Laptop","Mobile","Tablet",
        "Laptop","Mobile","Tablet","Laptop","Mobile"
    ],
    "Category": [
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics"
    ],
    "City": [
        "Delhi","Mumbai","Lucknow","Delhi","Lucknow",
        "Mumbai","Delhi","Lucknow","Mumbai","Delhi",
        "Lucknow","Mumbai","Delhi","Lucknow","Mumbai",
        "Delhi","Lucknow","Mumbai","Delhi","Lucknow"
    ],
    "Sales": [
        65000,30000,25000,70000,32000,
        27000,68000,35000,26000,72000,
        33000,28000,75000,34000,29000,
        76000,36000,30000,78000,37000
    ],
    "Profit": [
        12000,5000,4500,13000,5500,
        4700,12500,6000,4800,14000,
        6200,5000,14500,6300,5200,
        15000,6500,5400,15500,6800
    ],
    "Quantity": [
        5,10,8,6,9,
        7,5,11,8,6,
        10,7,5,9,8,
        4,12,9,5,10
    ]
})

print(df)
```

```python
import plotly.express as px
fig=px.scatter(df,x='Profit',y='Sales',title='Profit vs Sales')
fig.show()
```

```python
# Bar Chart – Total Sales for each Product

sales_product = df.groupby("Product")["Sales"].sum().reset_index()
fig = px.bar(sales_product,x="Product",y="Sales",title="Total Sales by Product",color="Product")
fig.show()
```

```python
#Pie Chart – Sales Distribution by Product

sales_product = df.groupby("Product")["Sales"].sum().reset_index()
fig = px.pie(sales_product,names="Product",values="Sales",title="Sales Distribution by Product")
fig.show()
```

```python
#Scatter Plot – Sales vs Profit

fig = px.scatter(
    df,
    x="Sales",
    y="Profit",
    color="Product",
    size="Quantity",
    title="Sales vs Profit",
    hover_data=["City"]
)

fig.show()
```

```python
#Histogram – Sales (10 Bins)

fig = px.histogram(
    df,
    x="Sales",
    nbins=10,
    title="Histogram of Sales"
)

fig.show()
```

```python
#Box Plot – Sales Distribution of Each Product

fig = px.box(
    df,
    x="Product",
    y="Sales",
    color="Product",
    title="Sales Distribution by Product"
)

fig.show()
```

```python
#Bar Chart – Total Sales of Each City

city_sales = df.groupby("City")["Sales"].sum().reset_index()

fig = px.bar(
    city_sales,
    x="City",
    y="Sales",
    color="City",
    title="Total Sales by City"
)

fig.show()
```

```python
#Line Chart – Average Profit of Each Product

avg_profit = df.groupby("Product")["Profit"].mean().reset_index()
fig = px.line(
    avg_profit,
    x="Product",
    y="Profit",
    markers=True,
    title="Average Profit by Product"
)

fig.show()
```
