```python
import numpy as np

# Student Marks Dataset (out of 100)
marks = np.array([45, 67, 89, 90, 56, 78, 88, 92, 61, 74,
                  83, 95, 52, 68, 71, 85, 77, 64, 58, 81])

print("Dataset:")
print(marks)
```

```python
#mean
mean = np.mean(marks)
print(" Mean =", mean)
```

```python
#median
median = np.median(marks)
print(" Median =", median)
```

```python
#mode
values, counts = np.unique(marks, return_counts=True)
max_count = np.max(counts)
if max_count > 1:
    mode = values[counts == max_count]
    print(" Mode =", mode)
else:
    print("No mode (all values occur only once)")
```

```python
#Maximum and Minimum
print(" Maximum =", np.max(marks))
print("   Minimum =", np.min(marks))
```

```python
#range
data_range = np.max(marks) - np.min(marks)
print("Range =", data_range)
```

```python
# Variance
variance = np.var(marks)
print(" Variance =", variance)
```

```python
#Standard Deviation
std_dev = np.std(marks)
print( "Standard Deviation =", std_dev)
```

```python
# 25th Percentile (Q1)
q1 = np.percentile(marks, 25)
print(" Q1 (25th Percentile) =", q1)
```

```python
#50th Percentile (Median)
q2 = np.percentile(marks, 50)
print(" Q2 (Median) =", q2)
```

```python
#75th Percentile (Q3)
q3 = np.percentile(marks, 75)
print(" Q3 (75th Percentile) =", q3)
```

```python
#Interquartile Range (IQR)
iqr = q3 - q1
print(" IQR =", iqr)
```

```python
#Count students scoring above the mean
above_mean = np.sum(marks > mean)
print(" Students scoring above mean =", above_mean)
```

```python
#Count students scoring below 60
below_60 = np.sum(marks < 60)
print(" Students scoring below 60 =", below_60)
```

```python
#Sum and Manual Average
total = np.sum(marks)
average = total / len(marks)
print(" Total Marks =", total)
print("    Manual Average =", average)
```

```python
#Sort Ascending, Descending and Top 5 Scores
ascending = np.sort(marks)
descending = ascending[::-1]
top5 = descending[:5]

print("Ascending Order =", ascending)
print("    Descending Order =", descending)
print("    Top 5 Highest Scores =", top5)
```
