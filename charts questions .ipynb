```python
#Q17. Plot two lines on the same graph: Average Marks by Course ,Maximum Marks by Course


course = df.groupby("Course")["Marks"]
plt.figure(figsize=(8,5))
plt.plot(course.mean().index, course.mean().values, marker='o', label="Average Marks")
plt.plot(course.max().index, course.max().values, marker='s', label="Maximum Marks")
plt.xlabel("Course")
plt.ylabel("Marks")
plt.title("Average and Maximum Marks by Course")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#18. Create a grouped bar chart comparing: Average Marks, Maximum Marks for each Course.

import numpy as np
course = df.groupby("Course")["Marks"]
avg = course.mean()
mx = course.max()
x = np.arange(len(avg))
w = 0.35
plt.figure(figsize=(8,5))
plt.bar(x-w/2, avg.values, width=w, label="Average")
plt.bar(x+w/2, mx.values, width=w, label="Maximum")
plt.xticks(x, avg.index)
plt.xlabel("Course")
plt.ylabel("Marks")
plt.title("Average vs Maximum Marks")
plt.legend()
plt.show()
```

```python
#19. Create a bar chart of the Top 10 students by Marks, coloring bars: Green if Marks ≥ 85 Red otherwise

top = df.nlargest(10, "Marks")
colors = ["red" if i >= 85 else "red" for i in top["Marks"]]
plt.figure(figsize=(9,5))
plt.bar(top["Name"], top["Marks"], color=colors)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Top 10 Students")
plt.xticks(rotation=45)
plt.show()
```
